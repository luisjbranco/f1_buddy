```python
circuits = pd.read_csv("data/wc_1950_2022/circuits.csv")
constructor_results = pd.read_csv("data/wc_1950_2022/constructor_results.csv")
constructors = pd.read_csv("data/wc_1950_2022/constructors.csv")
constructor_standings = pd.read_csv("data/wc_1950_2022/constructor_standings.csv")
drivers = pd.read_csv("data/wc_1950_2022/drivers.csv")
driver_standings = pd.read_csv("data/wc_1950_2022/driver_standings.csv")
lap_times = pd.read_csv("data/wc_1950_2022/lap_times.csv")
pit_stops = pd.read_csv("data/wc_1950_2022/pit_stops.csv")
qualifying = pd.read_csv("data/wc_1950_2022/qualifying.csv")
races = pd.read_csv("data/wc_1950_2022/races.csv")
results = pd.read_csv("data/wc_1950_2022/results.csv")
seasons = pd.read_csv("data/wc_1950_2022/seasons.csv")
sprint_results = pd.read_csv("data/wc_1950_2022/sprint_results.csv")
status = pd.read_csv("data/wc_1950_2022/status.csv")
```
